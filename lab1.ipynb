# Numerical Differentiation with the Finite Difference Method

The finite difference method is a numerical technique for approximating the derivative of a function. It is based on the idea that the derivative of a function can be approximated by the slope of a line tangent to the function at a point. The slope of the line is given by the difference in the function values at two points divided by the difference in the x-coordinates of the two points. This is the basic idea behind the finite difference method.

# Task 1

## Forward Difference Method

$$f'(x) = \frac{f(x+h) - f(x)}{h}$$

## Backward Difference Method

$$f'(x) = \frac{f(x) - f(x-h)}{h}$$

## Central Difference Method

$$f'(x) = \frac{f(x+h) - f(x-h)}{2h}$$

## Order of the approximation of the central difference method

The central difference method is more accurate than the forward and backward difference methods. The error in the central difference method is proportional to $h^2$, where $h$ is the step size. This means that the central difference method is a second-order approximation of the derivative.

### Proof
Given the Taylor series expansion of $f(x+h)$ and $f(x-h)$, we can write

$$f(x+h) = f(x) + hf'(x) + \frac{h^2}{2}f''(x) + \frac{h^3}{6}f'''(x) + \frac{h^4}{24}f''''(x) + \cdots$$

$$f(x-h) = f(x) - hf'(x) + \frac{h^2}{2}f''(x) - \frac{h^3}{6}f'''(x) + \frac{h^4}{24}f''''(x) - \cdots$$

Subtracting the second equation from the first, we get

$$f(x+h) - f(x-h) = 2hf'(x) + \frac{h^3}{3}f'''(x) + \frac{h^5}{60}f''''(x) + \cdots$$

Solving for $f'(x)$, we get

$$f'(x) = \frac{f(x+h) - f(x-h)}{2h} - \frac{h^2}{6}f'''(x) - \frac{h^4}{120}f''''(x) - \cdots$$

The error in the central difference method is proportional to $h^2$, which means that the central difference method is a second-order approximation of the derivative.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def forward_diff(f, x, h=1e-4):
    return (f(x + h) - f(x)) / h

def backward_diff(f, x, h=1e-4):
    return (f(x) - f(x - h)) / h

def central_diff(f, x, h=1e-4):
    return (f(x + h) - f(x - h)) / (2 * h)
```

```python
def f(x):
    return np.cos(x)

x = np.linspace(0, 2 * np.pi, 100)
y = f(x)

plt.plot(x, y, label='cos(x)')
plt.plot(x, -np.sin(x), label='-sin(x)')
plt.plot(x, forward_diff(f, x), label='forward')
plt.plot(x, backward_diff(f, x), label='backward')
plt.plot(x, central_diff(f, x), label='central')
plt.title('Differentiation of cos(x) h=1e-4')
plt.legend()
plt.show()
```

```python
x = 1
print("h=1e-4")
print('f\'(x) =', -np.sin(x))
print('forward:', forward_diff(f, x,1e-4) ,'Error: ', abs(forward_diff(f, x,1e-4) - (-np.sin(x))))
print('backward:', backward_diff(f, x, 1e-4), 'Error: ', abs(backward_diff(f, x, 1e-4) - (-np.sin(x))))
print('central:', central_diff(f, x,1e-4), 'Error: ', abs(central_diff(f, x,1e-4) - (-np.sin(x))))
```

```python
def finite_difference(f, x, h):
    # Calculate the values of the function at x, x+h, and x+3h
    u_j = f(x)
    u_j1 = f(x + h)
    u_j3 = f(x + 3*h)

    # Forward difference
    forward_diff = (u_j1 - u_j) / h

    # Central difference
    central_diff = (u_j3 - u_j) / (2 * h)

    # Combine the two approximations
    u_j_prime = (forward_diff + central_diff) / 2

    return u_j_prime
```
